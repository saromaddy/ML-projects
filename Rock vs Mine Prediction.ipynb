```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
#Loading the dataset
sonar_data = pd.read_csv('Copy of sonar data.csv', header=None)
```

```python
sonar_data.head()
```

```python
# for binary classification Logistic regression works well
# Supervised Learning

sonar_data.shape
```

```python
sonar_data.describe()
```

```python
sonar_data[60].value_counts()
#R - Rock
#M - Mine
```

```python
sonar_data.groupby(60).mean()
```

```python
# seperate the target variable (data and label)
x = sonar_data.drop(columns=60, axis=1)
y = sonar_data[60]
```

```python
x_train, x_test ,y_train, y_test = train_test_split(x,y, test_size=0.1, stratify=y,random_state=1)
# order of this is important
# test_size is 10% of the entire data set
# stratify means based on that y which means rock and mine are equal in training data preserving proportion
# random_state = this is to get the same data set for train test split each time when i run this code.
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
#model training
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#model evaluation
# accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
print('Accuracy on training data', training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
print('Accuracy on testing data',test_data_accuracy)
```
