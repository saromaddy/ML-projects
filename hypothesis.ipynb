```python
# ðŸ“Œ Hypothesis Testing with Z-Test and T-Test
# Import required libraries
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style="whitegrid")
```

```python
# Z-Test setup
sample_mean = 351
population_mean = 355
std_dev = 10
n = 50

z = (sample_mean - population_mean) / (std_dev / np.sqrt(n))
p_value = stats.norm.cdf(z)

# Plot the normal distribution
x = np.linspace(population_mean - 4*std_dev, population_mean + 4*std_dev, 1000)
y = stats.norm.pdf(x, population_mean, std_dev / np.sqrt(n))

plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Sampling Distribution (Î¼ = 355)")
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Sample Mean = {sample_mean}")
plt.fill_between(x, 0, y, where=(x <= sample_mean), color='red', alpha=0.3)
plt.title("Z-Test Visualization")
plt.xlabel("Sample Means")
plt.ylabel("Probability Density")
plt.legend()
plt.show()

print(f"Z-score = {z:.2f}, P-value = {p_value:.4f}")
```

```python
sample_scores = [70, 78, 74, 72, 68, 76, 73, 75, 77, 74, 71, 69]
sample_mean = np.mean(sample_scores)
t_stat, p_value = stats.ttest_1samp(sample_scores, 75)

# Plotting
sns.histplot(sample_scores, kde=True, bins=8, color='skyblue')
plt.axvline(75, color='red', linestyle='--', label="Population Mean = 75")
plt.axvline(sample_mean, color='green', linestyle='--', label=f"Sample Mean = {sample_mean:.2f}")
plt.title("One Sample T-Test: Distribution of Student Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.legend()
plt.show()

print(f"T-statistic = {t_stat:.2f}, P-value = {p_value:.4f}")
```

```python
male_scores = [65, 70, 72, 68, 75, 78, 74, 77]
female_scores = [72, 74, 69, 80, 73, 77, 75, 79]

# Combine for plotting
df = pd.DataFrame({
    'Scores': male_scores + female_scores,
    'Gender': ['Male']*len(male_scores) + ['Female']*len(female_scores)
})

sns.boxplot(x="Gender", y="Scores", data=df, palette='Set2')
plt.title("Two Sample T-Test: Male vs Female Math Scores")
plt.show()

# Run T-test
t_stat, p_value = stats.ttest_ind(male_scores, female_scores)
print(f"T-statistic = {t_stat:.2f}, P-value = {p_value:.4f}")
```

```python
# Math scores
male_scores = [65, 70, 72, 68, 75, 78, 74, 77]
female_scores = [72, 74, 69, 80, 73, 77, 75, 79]

# T-test
t_stat, p_value = stats.ttest_ind(male_scores, female_scores)

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There's a significant difference in scores between genders.")
else:
    print("Fail to reject the null hypothesis: No significant difference in scores.")

```
